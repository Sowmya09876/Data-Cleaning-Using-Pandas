```python
### Replacing null values with median ###
```

```python
Avg_age = data['Age'].mean()
Avg_age
```

```python
Median = data['Age'].median()
Median
```

```python
data['Age'] = data.Age.fillna(Median)
```

```python
data.head()
```

```python
avg_rating = data['Rating'].mean()
avg_rating
```

```python
data['Rating'] = data.Rating.fillna(avg_rating)
data['Rating'] = data.Rating.round(decimals=1)
data
```

```python
###Salary column###
```

```python
data['Salary'] = data.Salary.str.replace("k","000")
data
```

```python
### Location ###
```

```python
data['Location'] = data['Location'].str.split(",",expand=True)[0]
data
```

```python
data['Location'] = data['Location'].replace({"India In":"India","Australia Aus":"Australia"})
data
```

```python
data['Easy Apply'] = data['Easy Apply'].replace("-1","TRUE")
data
```

```python
data['Established'] = data['Established'].replace(-1,"NA")
data
```
